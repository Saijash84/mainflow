1.How many people have heart disease and how many people doesn't have heart disease?

```python
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind='bar',color=['green','red'])
plt.title('Heart Disease Distribution')
plt.xlabel('1=Heart Disease, 0=No heart disease')
plt.ylabel('Count')
plt.show()
```

```python
#plotting a pie chart
df.target.value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.legend("Disease","No Disease")
plt.title('Heart Disease Distribution')
plt.show()
```

526 people have heart disease whereas 499 people doesn't

2.People of which sex has most heart diseases?

```python
df.sex.value_counts()
```

```python
pd.crosstab(df.sex,df.target)
```

```python
sns.countplot(x='sex',hue='target',data=df)
plt.title('Heart Disease Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

Male have more heart disease

3.People of which sex has which type of chest pain most?

```python
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar')
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python
x=pd.crosstab(df.sex,df.cp)
x
```

```python
plt.figure(figsize=(8, 6))
x.plot(kind='bar',stacked=True)
plt.title('Chest Pain Type Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

Both female and male have 0 type chest pain most


```python
sns.displot(x='age',data=df,bins=30,kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

4.People with which chest pain are most pron to have heart disease?

```python
pd.crosstab(df.cp,df.target)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='cp',hue='target',data=df)
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

5)How does the resting blood pressure (trestbps) vary among individuals with and without heart disease?

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='trestbps', data=df)
plt.title('Resting Blood Pressure and Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

Resting blood pressure is more for those who doesn't habve heart disease

6)How does thalach varies with different chest pain?

```python
sns.barplot(x='cp',y='thalach',data=df)
plt.title('Chest pain vs thalach')
plt.xlabel('Chest Pain Type')
plt.ylabel('thalach')
plt.show()
```

Thalach doesn't effect the chest pain more

7.Derive the relationship between gender and slope?
